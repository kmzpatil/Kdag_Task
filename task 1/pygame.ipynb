````python
import pygame
import numpy as np
import random
pygame.init()
direction = {0:'UP', 1:'LEFT', 2:'DOWN', 3:'RIGHT'}
box_size = 15
height = 720
width = 1080
rows = height // box_size
cols = width // box_size
````

````python
class Ant():
    def __init__(self,x,y,color,face,perfume):
        self.x = x
        self.y = y
        self.color = color
        self.face = face
        self.perfume = perfume
        
    
    def move_ant(self):
        if direction[self.face] == "UP":
            self.y += box_size
        elif direction[self.face] == "DOWN":
            self.y -= box_size
        elif direction[self.face] == "LEFT":
            self.x -= box_size
        elif direction[self.face] == "RIGHT":
            self.x += box_size
        
    def turn_left(self):
        self.face = (self.face+3)%4
    
    def turn_right(self):
        self.face = (self.face+1)%4
````

````python
screen = pygame.display.set_mode((width,height))

grid_color = np.zeros((width, height))
grid_perfume = np.zeros((width, height,2))


ant1 = Ant(width//2+box_size,height//2,(255,0,0),0,1)
ant2 = Ant(width//2-box_size,height//2,(0,0,255),0,1)
running = True

while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
    
    
    
    
    screen.fill((255,255,255))
    for i in range(cols):
        for j in range(rows):
            if grid_color[i][j] == 0:
                color = (255,255,255)
            elif grid_color[i][j] == 1:
                color = (0,0,0)
            pygame.draw.rect(screen,color,(i*box_size,j*box_size,box_size,box_size))
            pygame.draw.rect(screen,(150,150,150),(i*box_size,j*box_size,box_size,box_size),1)
            
            
       
            
            
    pygame.draw.rect(screen,ant1.color,(ant1.x,ant1.y,box_size,box_size))    
    pygame.draw.rect(screen,ant2.color,(ant2.x,ant2.y,box_size,box_size)) 
    
    
    
    grid_color[ant1.x//box_size][ant1.y//box_size] = 1-grid_color[ant1.x//box_size][ant1.y//box_size]
    grid_color[ant2.x//box_size][ant2.y//box_size] = 1-grid_color[ant2.x//box_size][ant2.y//box_size]
    
    
    
    
    
    if(grid_perfume[ant1.x//box_size][ant1.y//box_size][0]==0):
        if grid_color[ant1.x//box_size][ant1.y//box_size] == 0:
            ant1.turn_right()
        else:
            ant1.turn_left()
    
    if(grid_perfume[ant1.x//box_size][ant1.y// box_size][0]==1):
        a = random.random()>0.8*grid_perfume[i][j][1]/5
        if a:
            if grid_color[ant1.x//box_size][ant1.y//box_size] == 0:
                ant1.turn_right()
            else:
                ant1.turn_left()
    
    if(grid_perfume[ant1.x//box_size][ant1.y// box_size][0]==2):
        a = random.random()>1-0.8*grid_perfume[i][j][1]/5
        if a:
            if grid_color[ant1.x//box_size][ant1.y//box_size] == 0:
                ant1.turn_right()
            else:
                ant1.turn_left()
    
    
    
    
    
    
    
    if(grid_perfume[ant2.x//box_size][ant2.y//box_size][0]==0):
        if grid_color[ant2.x//box_size][ant2.y//box_size] == 0:
            ant2.turn_right()
        else:
            ant2.turn_left()
    
    if(grid_perfume[ant2.x//box_size][ant2.y// box_size][0]==2):
        a = random.random()>0.8*grid_perfume[i][j][1]/5
        if a:
            if grid_color[ant2.x//box_size][ant2.y//box_size] == 0:
                ant2.turn_right()
            else:
                ant2.turn_left()        
    
    if(grid_perfume[ant2.x//box_size][ant2.y// box_size][0]==1):
        a = random.random()>1-0.8*grid_perfume[i][j][1]/5
        if a:
            if grid_color[ant2.x//box_size][ant2.y//box_size] == 0:
                ant2.turn_right()
            else:
                ant2.turn_left()            
                
                
    
    for i in range(cols):
        for j in range(rows):
            if (grid_perfume[i][j][1]>0):
                grid_perfume[i][j][1] -= 1 
              
                
    
    
    
    
    grid_perfume[ant1.x//box_size][ant1.y//box_size] = [1,5]
    grid_perfume[ant2.x//box_size][ant2.y//box_size]= [2,5]
    
    
    ant1.move_ant()
    ant2.move_ant()
        
        
        
    pygame.display.flip()
    pygame.time.delay(10)
    


pygame.quit()
````

````python
import pygame
import random

# Initialize PyGame
pygame.init()

# Constants
WIDTH, HEIGHT = 600, 600
GRID_SIZE = 20  # Each cell is 20x20 pixels
ROWS, COLS = HEIGHT // GRID_SIZE, WIDTH // GRID_SIZE
WHITE, BLACK, RED, BLUE = (255, 255, 255), (0, 0, 0), (255, 0, 0), (0, 0, 255)

# Create the game window
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Langton's Ants with Pheromones")

# Grid (0 = white, 1 = black)
grid = [[0 for _ in range(COLS)] for _ in range(ROWS)]
pheromones = [[{"A": 0, "B": 0} for _ in range(COLS)] for _ in range(ROWS)]  # Pheromone storage

# Define Ant class
class Ant:
    def __init__(self, x, y, color, pheromone_type):
        self.x = x
        self.y = y
        self.color = color
        self.pheromone_type = pheromone_type
        self.directions = ["UP", "RIGHT", "DOWN", "LEFT"]
        self.current_direction = 0  # Start facing UP

    def turn_right(self):
        self.current_direction = (self.current_direction + 1) % 4

    def turn_left(self):
        self.current_direction = (self.current_direction - 1) % 4

    def move_forward(self):
        if self.directions[self.current_direction] == "UP" and self.y > 0:
            self.y -= 1
        elif self.directions[self.current_direction] == "DOWN" and self.y < ROWS - 1:
            self.y += 1
        elif self.directions[self.current_direction] == "LEFT" and self.x > 0:
            self.x -= 1
        elif self.directions[self.current_direction] == "RIGHT" and self.x < COLS - 1:
            self.x += 1

    def update(self):
        cell_color = grid[self.y][self.x]
        cell_pheromones = pheromones[self.y][self.x]

        # Self-pheromone effect: 80% chance to move straight
        if cell_pheromones[self.pheromone_type] > 0:
            if random.random() < 0.8:
                self.move_forward()
                return

        # Cross-pheromone effect: 80% chance to follow normal rules
        other_pheromone = "A" if self.pheromone_type == "B" else "B"
        if cell_pheromones[other_pheromone] > 0:
            if random.random() > 0.8:
                self.move_forward()
                return

        # Apply Langtonâ€™s Ant rules
        if cell_color == 0:  # White square
            self.turn_right()
        else:  # Black square
            self.turn_left()

        grid[self.y][self.x] = 1 - grid[self.y][self.x]  # Flip color

        # Update pheromones
        pheromones[self.y][self.x] = {"A": 0, "B": 0}  # Remove old pheromones
        pheromones[self.y][self.x][self.pheromone_type] = 5  # Set new pheromone

        self.move_forward()

# Create two ants
ant1 = Ant(COLS // 3, ROWS // 2, RED, "A")
ant2 = Ant(2 * COLS // 3, ROWS // 2, BLUE, "B")

ants = [ant1, ant2]

def draw_grid():
    """Draws the grid with filled colors."""
    for row in range(ROWS):
        for col in range(COLS):
            color = BLACK if grid[row][col] == 1 else WHITE
            pygame.draw.rect(screen, color, (col * GRID_SIZE, row * GRID_SIZE, GRID_SIZE, GRID_SIZE))
            pygame.draw.rect(screen, (200, 200, 200), (col * GRID_SIZE, row * GRID_SIZE, GRID_SIZE, GRID_SIZE), 1)

def draw_ants():
    """Draws all ants."""
    for ant in ants:
        pygame.draw.rect(screen, ant.color, (ant.x * GRID_SIZE, ant.y * GRID_SIZE, GRID_SIZE, GRID_SIZE))

def decay_pheromones():
    """Reduces pheromone strength over time."""
    for row in range(ROWS):
        for col in range(COLS):
            for p in ["A", "B"]:
                if pheromones[row][col][p] > 0:
                    pheromones[row][col][p] -= 1

# Game loop
running = True
while running:
    screen.fill(WHITE)
    draw_grid()
    draw_ants()

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    for ant in ants:
        ant.update()

    decay_pheromones()  # Apply pheromone decay

    pygame.display.flip()
     # Adjust speed

pygame.quit()
````
