````python
import pygame

# Initialize PyGame
pygame.init()

# Constants
WIDTH, HEIGHT = 600, 600
GRID_SIZE = 20  # Each cell is 20x20 pixels
ROWS, COLS = HEIGHT // GRID_SIZE, WIDTH // GRID_SIZE
WHITE, BLACK, RED = (255, 255, 255), (0, 0, 0), (255, 0, 0)

# Create the game window
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Langton's Ant")

# Grid (0 = white, 1 = black)
grid = [[0 for _ in range(COLS)] for _ in range(ROWS)]

# Ant properties
ant_x, ant_y = COLS // 2, ROWS // 2  # Start in the center
directions = ["UP", "RIGHT", "DOWN", "LEFT"]
current_direction = 0  # Start facing UP

def draw_grid():
    """Draws the grid with filled colors."""
    for row in range(ROWS):
        for col in range(COLS):
            color = BLACK if grid[row][col] == 1 else WHITE
            pygame.draw.rect(screen, color, (col * GRID_SIZE, row * GRID_SIZE, GRID_SIZE, GRID_SIZE))
            pygame.draw.rect(screen, (200, 200, 200), (col * GRID_SIZE, row * GRID_SIZE, GRID_SIZE, GRID_SIZE), 1)

def draw_ant():
    """Draws the ant."""
    pygame.draw.rect(screen, RED, (ant_x * GRID_SIZE, ant_y * GRID_SIZE, GRID_SIZE, GRID_SIZE))

def turn_right():
    """Turns the ant 90 degrees clockwise."""
    global current_direction
    current_direction = (current_direction + 1) % 4  # Cycle through directions

def turn_left():
    """Turns the ant 90 degrees counterclockwise."""
    global current_direction
    current_direction = (current_direction - 1) % 4  # Cycle through directions

def move_forward():
    """Moves the ant forward in the current direction."""
    global ant_x, ant_y
    if directions[current_direction] == "UP" and ant_y > 0:
        ant_y -= 1
    elif directions[current_direction] == "DOWN" and ant_y < ROWS - 1:
        ant_y += 1
    elif directions[current_direction] == "LEFT" and ant_x > 0:
        ant_x -= 1
    elif directions[current_direction] == "RIGHT" and ant_x < COLS - 1:
        ant_x += 1

def update_ant():
    """Applies Langton's Ant movement rules."""
    global grid
    if grid[ant_y][ant_x] == 0:  # White square
        turn_right()
    else:  # Black square
        turn_left()
    grid[ant_y][ant_x] = 1 - grid[ant_y][ant_x]  # Flip color
    move_forward()

# Game loop
running = True
while running:
    screen.fill(WHITE)
    draw_grid()
    draw_ant()

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    update_ant()  # Move the ant according to Langton's rules
    pygame.display.flip()  # Adjust speed of simulation

pygame.quit()
````

````python
import pygame
import random

# Initialize PyGame
pygame.init()

# Constants
WIDTH, HEIGHT = 600, 600
GRID_SIZE = 20  # Each cell is 20x20 pixels
ROWS, COLS = HEIGHT // GRID_SIZE, WIDTH // GRID_SIZE
WHITE, BLACK, RED, BLUE = (255, 255, 255), (0, 0, 0), (255, 0, 0), (0, 0, 255)

# Create the game window
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Langton's Ants with Pheromones")

# Grid (0 = white, 1 = black)
grid = [[0 for _ in range(COLS)] for _ in range(ROWS)]
pheromones = [[{"A": 0, "B": 0} for _ in range(COLS)] for _ in range(ROWS)]  # Pheromone storage

# Define Ant class
class Ant:
    def __init__(self, x, y, color, pheromone_type):
        self.x = x
        self.y = y
        self.color = color
        self.pheromone_type = pheromone_type
        self.directions = ["UP", "RIGHT", "DOWN", "LEFT"]
        self.current_direction = 0  # Start facing UP

    def turn_right(self):
        self.current_direction = (self.current_direction + 1) % 4

    def turn_left(self):
        self.current_direction = (self.current_direction - 1) % 4

    def move_forward(self):
        if self.directions[self.current_direction] == "UP" and self.y > 0:
            self.y -= 1
        elif self.directions[self.current_direction] == "DOWN" and self.y < ROWS - 1:
            self.y += 1
        elif self.directions[self.current_direction] == "LEFT" and self.x > 0:
            self.x -= 1
        elif self.directions[self.current_direction] == "RIGHT" and self.x < COLS - 1:
            self.x += 1

    def update(self):
        cell_color = grid[self.y][self.x]
        cell_pheromones = pheromones[self.y][self.x]

        # Self-pheromone effect: 80% chance to move straight
        if cell_pheromones[self.pheromone_type] > 0:
            if random.random() < 0.8:
                self.move_forward()
                return

        # Cross-pheromone effect: 80% chance to follow normal rules
        other_pheromone = "A" if self.pheromone_type == "B" else "B"
        if cell_pheromones[other_pheromone] > 0:
            if random.random() > 0.8:
                self.move_forward()
                return

        # Apply Langton’s Ant rules
        if cell_color == 0:  # White square
            self.turn_right()
        else:  # Black square
            self.turn_left()

        grid[self.y][self.x] = 1 - grid[self.y][self.x]  # Flip color

        # Update pheromones
        pheromones[self.y][self.x] = {"A": 0, "B": 0}  # Remove old pheromones
        pheromones[self.y][self.x][self.pheromone_type] = 5  # Set new pheromone

        self.move_forward()

# Create two ants
ant1 = Ant(COLS // 3, ROWS // 2, RED, "A")
ant2 = Ant(2 * COLS // 3, ROWS // 2, BLUE, "B")

ants = [ant1, ant2]

def draw_grid():
    """Draws the grid with filled colors."""
    for row in range(ROWS):
        for col in range(COLS):
            color = BLACK if grid[row][col] == 1 else WHITE
            pygame.draw.rect(screen, color, (col * GRID_SIZE, row * GRID_SIZE, GRID_SIZE, GRID_SIZE))
            pygame.draw.rect(screen, (200, 200, 200), (col * GRID_SIZE, row * GRID_SIZE, GRID_SIZE, GRID_SIZE), 1)

def draw_ants():
    """Draws all ants."""
    for ant in ants:
        pygame.draw.rect(screen, ant.color, (ant.x * GRID_SIZE, ant.y * GRID_SIZE, GRID_SIZE, GRID_SIZE))

def decay_pheromones():
    """Reduces pheromone strength over time."""
    for row in range(ROWS):
        for col in range(COLS):
            for p in ["A", "B"]:
                if pheromones[row][col][p] > 0:
                    pheromones[row][col][p] -= 1

# Game loop
running = True
while running:
    screen.fill(WHITE)
    draw_grid()
    draw_ants()

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    for ant in ants:
        ant.update()

    decay_pheromones()  # Apply pheromone decay

    pygame.display.flip()
     # Adjust speed

pygame.quit()
````

````python
import pygame
import numpy as np
import random

# Initialize PyGame
pygame.init()

# Constants
WIDTH, HEIGHT = 700, 600
GRID_SIZE = 10  # Smaller grid size for better performance
ROWS, COLS = HEIGHT // GRID_SIZE, WIDTH // GRID_SIZE
WHITE, BLACK, RED, BLUE, GRAY = (255, 255, 255), (0, 0, 0), (255, 0, 0), (0, 0, 255), (200, 200, 200)

# Create the game window
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Optimized Langton's Ants")

# NumPy-based Grid & Pheromones
grid = np.zeros((ROWS, COLS), dtype=int)
pheromones = np.zeros((ROWS, COLS, 2), dtype=int)  # Two pheromone layers (A=0, B=1)

# Simulation control
running_simulation = False

# Create a surface for batch drawing
grid_surface = pygame.Surface((WIDTH, HEIGHT))

class Ant:
    def __init__(self, x, y, color, pheromone_index):
        self.x = x
        self.y = y
        self.color = color
        self.pheromone_index = pheromone_index
        self.directions = np.array([[0, -1], [1, 0], [0, 1], [-1, 0]])  # UP, RIGHT, DOWN, LEFT
        self.current_direction = 0

    def move(self):
        """Moves the ant based on Langton's rules & pheromones."""
        global grid, pheromones

        cell_color = grid[self.y, self.x]
        pheromone_here = pheromones[self.y, self.x]

        # Self-pheromone effect (80% chance to move straight)
        if pheromone_here[self.pheromone_index] > 0 and random.random() < 0.8:
            pass  # Move straight
        elif pheromone_here[1 - self.pheromone_index] > 0 and random.random() > 0.8:
            pass  # Move straight
        else:
            # Turn based on color
            self.current_direction = (self.current_direction + (1 if cell_color == 0 else -1)) % 4

        # Flip the color of the cell
        grid[self.y, self.x] ^= 1  # Toggle 0/1 (white/black)

        # Update pheromones
        pheromones[self.y, self.x] = [0, 0]  # Reset pheromones
        pheromones[self.y, self.x, self.pheromone_index] = 5  # Apply new pheromone

        # Move forward
        move_vector = self.directions[self.current_direction]
        self.x = (self.x + move_vector[0]) % COLS
        self.y = (self.y + move_vector[1]) % ROWS

ants = [Ant(COLS // 3, ROWS // 2, RED, 0), Ant(2 * COLS // 3, ROWS // 2, BLUE, 1)]

def update_grid_surface():
    """Redraws the entire grid efficiently on a separate surface."""
    grid_surface.fill(WHITE)

    for y in range(ROWS):
        for x in range(COLS):
            if grid[y, x] == 1:
                pygame.draw.rect(grid_surface, BLACK, (x * GRID_SIZE, y * GRID_SIZE, GRID_SIZE, GRID_SIZE))

            # Pheromone visualization
            pheromone_a, pheromone_b = pheromones[y, x]
            if pheromone_a > 0:
                pygame.draw.rect(grid_surface, (255, 100, 100), (x * GRID_SIZE, y * GRID_SIZE, GRID_SIZE, GRID_SIZE))
            if pheromone_b > 0:
                pygame.draw.rect(grid_surface, (100, 100, 255), (x * GRID_SIZE, y * GRID_SIZE, GRID_SIZE, GRID_SIZE))
            if pheromone_a > 0 and pheromone_b > 0:
                pygame.draw.rect(grid_surface, (150, 50, 150), (x * GRID_SIZE, y * GRID_SIZE, GRID_SIZE, GRID_SIZE))

def decay_pheromones():
    """Reduces pheromone strength over time."""
    global pheromones
    pheromones = np.maximum(pheromones - 1, 0)  # Ensures values don’t go below 0

# Button class for UI controls
class Button:
    def __init__(self, x, y, width, height, text, action):
        self.rect = pygame.Rect(x, y, width, height)
        self.text = text
        self.action = action

    def draw(self):
        pygame.draw.rect(screen, (50, 50, 50), self.rect)
        font = pygame.font.Font(None, 30)
        text_surf = font.render(self.text, True, WHITE)
        screen.blit(text_surf, (self.rect.x + 10, self.rect.y + 10))

    def click(self, pos):
        if self.rect.collidepoint(pos):
            self.action()

# Button actions
def toggle_simulation():
    global running_simulation
    running_simulation = not running_simulation

def reset_simulation():
    global grid, pheromones, ants, running_simulation
    grid.fill(0)
    pheromones.fill(0)
    ants = [Ant(COLS // 3, ROWS // 2, RED, 0), Ant(2 * COLS // 3, ROWS // 2, BLUE, 1)]
    running_simulation = False
    update_grid_surface()  # Reset grid surface

# Create buttons
buttons = [
    Button(WIDTH - 180, 50, 150, 40, "Start/Stop", toggle_simulation),
    Button(WIDTH - 180, 110, 150, 40, "Reset", reset_simulation),
]

# Initial grid rendering
update_grid_surface()

# Game loop
running = True
while running:
    screen.fill(WHITE)
    screen.blit(grid_surface, (0, 0))

    # Draw ants
    for ant in ants:
        pygame.draw.rect(screen, ant.color, (ant.x * GRID_SIZE, ant.y * GRID_SIZE, GRID_SIZE, GRID_SIZE))

    # Event Handling
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        if event.type == pygame.MOUSEBUTTONDOWN:
            for button in buttons:
                button.click(event.pos)

    # Simulation Logic
    if running_simulation:
        for ant in ants:
            ant.move()
        decay_pheromones()
        update_grid_surface()  # Only redraw changed grid state

    # Draw Buttons
    for button in buttons:
        button.draw()

    pygame.display.flip()
    pygame.time.delay(50)  # Slight delay for smoother animation

pygame.quit()
````

````python
import pygame
import numpy as np
import random

# Initialize PyGame
pygame.init()

# Constants
WIDTH, HEIGHT = 800, 600
GRID_SIZE = 10
ROWS, COLS = HEIGHT // GRID_SIZE, WIDTH // GRID_SIZE
WHITE, BLACK, RED, BLUE = (255, 255, 255), (0, 0, 0), (255, 0, 0), (0, 0, 255)

# Create the game window
screen = pygame.display.set_mode((WIDTH + 200, HEIGHT))
pygame.display.set_caption("Langton's Ants - Optimized with Real-Time Controls")

# NumPy-based Grid & Pheromones
grid = np.zeros((ROWS, COLS), dtype=int)
pheromones = np.zeros((ROWS, COLS, 2), dtype=int)

# UI Variables
running_simulation = False
pheromone_decay = 1
ant_speed = 50
self_pheromone_prob = 0.8
cross_pheromone_prob = 0.2

# Ant Class
class Ant:
    def __init__(self, x, y, color, pheromone_index):
        self.x = x
        self.y = y
        self.color = color
        self.pheromone_index = pheromone_index
        self.directions = np.array([[0, -1], [1, 0], [0, 1], [-1, 0]])
        self.current_direction = 0

    def move(self):
        global grid, pheromones
        cell_color = grid[self.y, self.x]
        pheromone_here = pheromones[self.y, self.x]
        
        if pheromone_here[self.pheromone_index] > 0 and random.random() < self_pheromone_prob:
            pass
        elif pheromone_here[1 - self.pheromone_index] > 0 and random.random() > cross_pheromone_prob:
            pass
        else:
            self.current_direction = (self.current_direction + (1 if cell_color == 0 else -1)) % 4
        
        grid[self.y, self.x] ^= 1
        pheromones[self.y, self.x] = [0, 0]
        pheromones[self.y, self.x, self.pheromone_index] = 5
        
        move_vector = self.directions[self.current_direction]
        self.x = (self.x + move_vector[0]) % COLS
        self.y = (self.y + move_vector[1]) % ROWS

# Initialize Ants
ants = [Ant(COLS // 3, ROWS // 2, RED, 0), Ant(2 * COLS // 3, ROWS // 2, BLUE, 1)]

# UI Controls
class Button:
    def __init__(self, x, y, width, height, text, action):
        self.rect = pygame.Rect(x, y, width, height)
        self.text = text
        self.action = action

    def draw(self):
        pygame.draw.rect(screen, (50, 50, 50), self.rect)
        font = pygame.font.Font(None, 30)
        text_surf = font.render(self.text, True, WHITE)
        screen.blit(text_surf, (self.rect.x + 10, self.rect.y + 10))

    def click(self, pos):
        if self.rect.collidepoint(pos):
            self.action()

# Button Actions
def toggle_simulation():
    global running_simulation
    running_simulation = not running_simulation

def reset_simulation():
    global grid, pheromones, ants, running_simulation
    grid.fill(0)
    pheromones.fill(0)
    ants[:] = [Ant(COLS // 3, ROWS // 2, RED, 0), Ant(2 * COLS // 3, ROWS // 2, BLUE, 1)]
    running_simulation = False

def adjust_decay(value):
    global pheromone_decay
    pheromone_decay = int(value)

def adjust_speed(value):
    global ant_speed
    ant_speed = int(value)

def adjust_self_pheromone(value):
    global self_pheromone_prob
    self_pheromone_prob = float(value) / 100

def adjust_cross_pheromone(value):
    global cross_pheromone_prob
    cross_pheromone_prob = float(value) / 100

# Create Buttons
buttons = [
    Button(WIDTH + 20, 50, 150, 40, "Start/Stop", toggle_simulation),
    Button(WIDTH + 20, 110, 150, 40, "Reset", reset_simulation)
]

# Main Loop
running = True
while running:
    screen.fill((30, 30, 30))
    pygame.draw.rect(screen, WHITE, (0, 0, WIDTH, HEIGHT))
    
    for y in range(ROWS):
        for x in range(COLS):
            if grid[y, x] == 1:
                pygame.draw.rect(screen, BLACK, (x * GRID_SIZE, y * GRID_SIZE, GRID_SIZE, GRID_SIZE))
    
    for ant in ants:
        pygame.draw.rect(screen, ant.color, (ant.x * GRID_SIZE, ant.y * GRID_SIZE, GRID_SIZE, GRID_SIZE))
    
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        if event.type == pygame.MOUSEBUTTONDOWN:
            for button in buttons:
                button.click(event.pos)
    
    if running_simulation:
        for ant in ants:
            ant.move()
        pheromones = np.maximum(pheromones - pheromone_decay, 0)
    
    for button in buttons:
        button.draw()
    
    pygame.display.flip()
    pygame.time.delay(ant_speed)

pygame.quit()
````
